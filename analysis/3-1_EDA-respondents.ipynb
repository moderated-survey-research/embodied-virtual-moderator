### Education

```python
educationMapping = {
    'low': ['None completed', 'Primary education', 'Secondary education'],
    'medium': ['High school diploma', 'Technical/community college'],
    'high': ['Undergraduate degree', 'Graduate degree' , 'Doctoral degree']
}
respondents['educationLevel'] = respondents.education.apply(lambda x: 'low' if x in educationMapping['low'] 
                                                             else 'medium' if x in educationMapping['medium']
                                                             else 'high')
```

```python
respondents.groupby('variant').education.value_counts().sort_index().unstack()
```

```python
respondents.groupby('variant').educationLevel.value_counts().sort_index().unstack()
```
