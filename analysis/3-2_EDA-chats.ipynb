## Latency

```python
messages = pd.read_csv('data/messages.csv')
followUpChats['latency'] = 0.0
for i, row in followUpChats.iterrows():
    messagesTimes = messages[(messages.chatID==row.id) & (messages.role=='user') & (messages.section=='follow_up')].totalTime
    followUpChats.loc[followUpChats.id == row.id, 'latency'] = row.duration * 60 - messagesTimes.sum()
```

```python
followUpChats.groupby('variant').latency.mean()
```
